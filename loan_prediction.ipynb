```python
# Encode categorical columns using .astype('category').cat.codes
categorical_cols = ['Married/Single', 'House_Ownership', 'Car_Ownership', 'Profession', 'CITY', 'STATE']

for col in categorical_cols:
    cleandata[col] = cleandata[col].astype('category').cat.codes
```

```python
cleandata.head()
```

```python
cleandata.dtypes
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x=cleandata.drop(['Risk_Flag'], axis=1).values
```

```python
y = cleandata['Risk_Flag'].values
```

```python
x_train, x_test, y_train, y_test,=train_test_split(x,y, test_size=0.15, random_state=1000)
```

```python
models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "SVM": SVC(),
    "KNN": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB(),
    "ADA": AdaBoostClassifier(),
    "BC": BaggingClassifier()
}

# Train and predict using each model
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{name}: Accuracy = {accuracy:.6f}")
```

```python
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    classification = classification_report(y_test, y_pred)
    print(f"{name}\n{classification}\n")
```
