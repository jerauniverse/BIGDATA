```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
cleandata=Fraud_Detection.drop(['TransactionID', 'AccountID', 'TransactionDate', 'PreviousTransactionDate', 'DeviceID', 'IP Address', 'MerchantID'], axis=1)
```

```python
cleandata.head()
```

```python
categorical_cols = ['TransactionType', 'Location', 'Channel', 'CustomerOccupation']

for col in categorical_cols:
    cleandata[col] = cleandata[col].astype('category').cat.codes
```

```python
cleandata.head()
```

```python
cleandata.dtypes
```

```python
cleandata['LoginAttempts'] = np.where(cleandata['LoginAttempts'] >= 3, 1, 0)
```

```python
cleandata.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x=cleandata.drop(['LoginAttempts'], axis=1).values
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
```

```python
y = cleandata['LoginAttempts'].values
```

```python
x_train, x_test, y_train, y_test,=train_test_split(x,y, test_size=0.1, random_state=1000)
```

```python
models = {
    "Logistic Regression": LogisticRegression(solver='lbfgs', max_iter=2000, C=0.1),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "SVM": SVC(),
    "KNN": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB(),
    "ADA": AdaBoostClassifier(),
    "BC": BaggingClassifier()
}

# Train and predict using each model
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{name}: Accuracy = {accuracy:.6f}")
```

```python
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    classification = classification_report(y_test, y_pred, zero_division=0)
    print(f"{name}\n{classification}\n")
```
